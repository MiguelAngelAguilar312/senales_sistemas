# Diagramas de Bode

Ejemplo de funci√≥n de transferencia
   $$H(s)=\dfrac{1}{s^2+2\zeta \omega_0s + \omega_0^2}$$

```python
#NOTE importar bibliotecas
import numpy             as np
import matplotlib.pyplot as plt

from scipy import signal
```

```python
#definir funcion
def plot_bode(Hs,color='b'):
    w, mag, phas = signal.bode(Hs)
    plt.figure('Magnitud')
    plt.semilogx(w, mag, color,linewidth=3)
    plt.grid('on')
    plt.figure('Fase')
    plt.semilogx(w, phas, color, linewidth=3)
    plt.grid('on')
    
    return plt
```

```python
w  = np.linspace(.01,30,1000)
#w  = 2*np.pi*w

wo = 8
B  = 1.
```

Ejemplo  $$H(s)=\dfrac{1}{s^2+16s + 64}$$

```python
Aw = np.sqrt(  (wo**2 - w**2)**2 + (B*wo*w)**2 )

plt.semilogx( w, -20*np.log(Aw), 'r', linewidth=3 )
plt.show()
```

Ejemplo  $$H(s)=\dfrac{1}{s^2+(2*0.5)8s + 64}$$ $$H(s)=\dfrac{1}{s^2+(2*2)8s + 64}$$

```python
wo  = 8
B   = 1.
Bl  = .5
Bg  = 2.
Aw  = np.sqrt(  (wo**2 - w**2)**2 + (B *wo*w)**2 )
Awl = np.sqrt(  (wo**2 - w**2)**2 + (Bl*wo*w)**2 )
Awg = np.sqrt(  (wo**2 - w**2)**2 + (Bg*wo*w)**2 )

plt.semilogx( w, -20*np.log(Aw ), 'r', linewidth=3 )
plt.semilogx( w, -20*np.log(Awl), 'g', linewidth=3 )
plt.semilogx( w, -20*np.log(Awg), 'b', linewidth=3 )
plt.show()
```

```python
Tw = np.arctan(  (wo**2 - w**2)/(B*wo*w) )

plt.semilogx( w, Tw, 'g', linewidth=3 )
plt.show()
```

```python
Tw  = np.arctan(  (wo**2 - w**2)/(B*wo*w) )
Twl = np.arctan(  (wo**2 - w**2)/(Bl*wo*w) )
Twg = np.arctan(  (wo**2 - w**2)/(Bg*wo*w) )

plt.semilogx( w, Tw , 'k', linewidth=3 )
plt.semilogx( w, Twl, 'g', linewidth=3 )
plt.semilogx( w, Twg, 'r', linewidth=3 )
plt.show()
```
