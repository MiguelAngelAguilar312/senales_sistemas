```python
import numpy             as np
import scipy             as sp
import matplotlib.pyplot as plt

from scipy import signal
```

```python
#definir funcion
def plot_bode(Hs,color='b'):
    w, mag, phas = signal.bode(Hs)
    plt.figure('Magnitud')
    plt.semilogx(w, mag, color,linewidth=3)
    plt.grid('on')
    plt.figure('Fase')
    plt.semilogx(w, phas, color, linewidth=3)
    plt.grid('on')
    
    return plt
```

```python
num = [1]
den = [1, 2, 1]
Hs  = signal.TransferFunction(num, den)

p = np.poly1d( den )
print( np.poly1d(p) )

plt = plot_bode(Hs)
plt.show()


```

```python
num = [1]
den = [ 1, np.sqrt(2), 1 ]
Hs  = signal.TransferFunction(num, den)

p = np.poly1d( den )
print( np.poly1d(p) )

plt = plot_bode(Hs)
plt.show()
```
