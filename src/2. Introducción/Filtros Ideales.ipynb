# Filtros ideales

```python
#NOTE importar bibliotecas
import numpy             as np
import matplotlib.pyplot as plt

from scipy import signal
```

### Filtro Pasa Bajas Frecuencias Ideal

```python
mn, mx  = 0, 2.5
t       = np.linspace(mn, mx, 500)
A       = np.ones( t.shape[0] )
A[t>1.] = 0


#plot
plt.xticks(fontsize = 15); plt.yticks(fontsize = 15)
plt.xlim(mn, mx)         ; plt.ylim(mn, 1.4)
plt.xlabel('Frecuencia Normalizada [Hz]',fontsize=14)
plt.ylabel('Amplitud [dB]',fontsize=14)
plt.title('Filtro Pasa Bajas Frecuencias Ideal',fontsize=18)

#plot format
plt.annotate(
        'Banda de paso',
        xy=(0, 0), xytext=(0.1, 1.1), color='r', fontsize=13)
plt.annotate(
        'Banda de rechazo',
        xy=(0, 0), xytext=(1.1, 0.1), color='r', fontsize=13)
plt.annotate(
        r'$\omega_p$ / $\omega_c$',
        xy=(1.05, 1), arrowprops=dict(arrowstyle='->'), xytext=(1.5, 1.1), color='r', fontsize=18)

plt.plot(t, A, 'b',linewidth=3)
plt.plot(1,1,'go',markersize=18,alpha=.75)
plt.plot(1,1,'ko',markersize=10)

plt.show()
```

### Filtro Altas Frecuencias Ideal

```python
mn, mx  = 0, 2.5
t       = np.linspace(mn, mx, 500)
A       = np.zeros( t.shape[0] )
A[t>1.] = 1

#plot
plt.plot(t, A, 'b',linewidth=3)
plt.xticks(fontsize = 15); plt.yticks(fontsize = 15)
plt.xlim(mn, mx)         ; plt.ylim(mn, 1.2)
plt.xlabel('Frecuencia Normalizada [Hz/Hz]',fontsize=14)
plt.ylabel('Amplitud [dB]',fontsize=14)
plt.title('Filtro Pasa Altas Frecuencias Ideal',fontsize=18)

#plot format
plt.annotate(
        'Banda de paso',
        xy=(0, 0), xytext=(1.3, 1.05), color='r', fontsize=13)
plt.annotate(
        'Banda de rechazo',
        xy=(0, 0), xytext=(0.04, 0.1), color='r', fontsize=13)
plt.annotate(
        r'$\omega_p$ / $\omega_c$',
        xy=(.95, 1), arrowprops=dict(arrowstyle='->'), xytext=(0.1, .9), color='r', fontsize=18)

plt.plot(t, A, linewidth=3)
plt.plot(1,1,'go',markersize=18,alpha=.75)
plt.plot(1,1,'ko',markersize=10)
plt.show()
```

### Filtro Pasa Banda de Frecuencias Ideal

```python
mn, mx   = 0, 2.5
t        = np.linspace(mn, mx, 500)
A        = np.ones( t.shape[0] )
A[t<.5]  = 0
A[t>1.5] = 0

#plot
plt.plot(t, A, 'b', linewidth=3)
plt.xticks(fontsize = 15); plt.yticks(fontsize = 15)
plt.xlim(mn, mx)         ; plt.ylim(mn, 1.2)
plt.xlabel('Frecuencia Normalizada [Hz/Hz]',fontsize=14)
plt.ylabel('Amplitud [dB]',fontsize=14)
plt.title('Filtro Pasa Banda de Frecuencias Ideal',fontsize=18)

#plot format
plt.annotate(
        'Banda de paso',
        xy=(0, 0), xytext=(.65, 1.1), color='r', fontsize=13)
plt.annotate(
        'Banda de rechazo',
        xy=(.25, 0.01), arrowprops=dict(arrowstyle='->'), xytext=(.55, .15), color='r', fontsize=13)
plt.annotate(
        'Banda de rechazo',
        xy=( 2, 0.01), arrowprops=dict(arrowstyle='->'), xytext=(.55, .15), color='r', fontsize=13)
plt.annotate(
        r'$\omega_{p1}$ / $\omega_{c1}$',
        xy=(.54, .97), arrowprops=dict(arrowstyle='->'), xytext=(0.7, .7), color='r', fontsize=18)
plt.annotate(
        r'$\omega_{p2}$ / $\omega_{c2}$',
        xy=(1.54, .97), arrowprops=dict(arrowstyle='->'), xytext=(1.7, .7), color='r', fontsize=18)

plt.plot(t, A, linewidth=3)
plt.plot( .5,1,'go',markersize=18,alpha=.75)
plt.plot( .5,1,'ko',markersize=10)
plt.plot(1.5,1,'go',markersize=18,alpha=.75)
plt.plot(1.5,1,'ko',markersize=10)

plt.show()
```

### Filtro Rechaza Banda de Frecuencias Ideal

```python
mn, mx   = 0, 2.5
t        = np.linspace(mn, mx, 500)
A        = np.zeros( t.shape[0] )
A[t<.85]  = 1
A[t>1.15] = 1

#plot
plt.plot(t, A, 'b', linewidth=3)
plt.xticks(fontsize = 15); plt.yticks(fontsize = 15)
plt.xlim(mn, mx)         ; plt.ylim(mn, 1.2)
plt.xlabel('Frecuencia Normalizada [Hz/Hz]',fontsize=14)
plt.ylabel('Amplitud [dB]',fontsize=14)
plt.title('Filtro Rechaza Banda de Frecuencias Ideal',fontsize=18)

#plot format
#plot format
plt.annotate(
        'Banda de paso',
        xy=(0, 0), xytext=(.05, 1.05), color='r', fontsize=13)
plt.annotate(
        'Banda de paso',
        xy=(0, 0), xytext=(1.3, 1.05), color='r', fontsize=13)
plt.annotate(
        'Banda de rechazo',
        xy=(0, 0), xytext=(.97, .65), color='r', fontsize=13,rotation=90)
plt.annotate(
        r'$\omega_{p1}$ / $\omega_{c1}$',
        xy=(.84, .97), arrowprops=dict(arrowstyle='->'), xytext=(.1, .7), color='r', fontsize=18)
plt.annotate(
        r'$\omega_{p2}$ / $\omega_{c2}$',
        xy=(1.16, .97), arrowprops=dict(arrowstyle='->'), xytext=(1.4, .7), color='r', fontsize=18)

plt.plot(t, A, linewidth=3)
plt.plot( .85,1,'go',markersize=18,alpha=.75)
plt.plot( .85,1,'ko',markersize=10)
plt.plot(1.15,1,'go',markersize=18,alpha=.75)
plt.plot(1.15,1,'ko',markersize=10)
plt.show()
```
